```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_polar(amp, angle):
    r = [0,amp[0],0,amp[1],0,amp[2]/2,0]
    theta = [0,angle[0]* np.pi/180,0, angle[1]* np.pi/180,0,angle[2]* np.pi/180,0]
    plt.figure(figsize=(10,10),dpi=200)

    plt.rcParams.update({'font.size': 20})
    ax = plt.subplot(111, projection='polar')
    ##ax.plot(theta, r, color='r', marker='+',linewidth=3)
    print(theta);
    print(r);
    ax.plot(theta[0:2], r[0:2], 'ro-')
    ax.annotate('DUT Amplitude:'+str(amp[0])+' '+'DUT phase angle:'+str(angle[0]) ,
            xy=(theta[1], r[1]),  # theta, radius
            xytext=(0.01, 0.01),    # fraction, fraction
            textcoords='figure fraction',
            arrowprops=dict(facecolor='red', shrink=0.05),
            horizontalalignment='left',
            verticalalignment='bottom',
            )
    
    ax.plot(theta[2:4], r[2:4], 'b*-')
    ax.annotate('REF Amplitude:'+str(amp[1])+' '+'REF phase angle:'+str(angle[1]) ,
            xy=(theta[3], r[3]),  # theta, radius
            xytext=(0.02, 0.02),    # fraction, fraction
            textcoords='figure fraction',
            arrowprops=dict(facecolor='blue', shrink=0.05),
            horizontalalignment='left',
            verticalalignment='bottom',
            )
    
    
    ax.set_rmax((int(max(amp)/100)+1)*100)
    ax.grid(True)
    ax.set_title("AC test polar plot [DUT,REF,ERR*10000]", loc = 'left',va='top',fontsize=15)
```

```python
amp_60Hz_off_board = [165.9,166.94,289.67]
angle_60Hz_off_board =[63.82, 360-116.02, 360-115.54]
```

```python
plot_polar(amp_60Hz_off_board,angle_60Hz_off_board)
```
