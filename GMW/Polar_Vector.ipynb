```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#amp = [159.71,277.55,87.93]
#angle =[307.1, 127.6, 128.8]
#[DUT,REF,ERR]
amp_60Hz_off_board = [165.9,166.94,289.67]
angle_60Hz_off_board =[63.82, 360-116.02, 360-115.54]

def plot_polar(amp, angle):
    r = [0,amp[0],0,amp[1],0,amp[2]/2,0]
    theta = [0,angle[0]* np.pi/180,0, angle[1]* np.pi/180,0,angle[2]* np.pi/180,0]
    plt.figure(figsize=(12,12),dpi=200)

    plt.rcParams.update({'font.size': 30})
    ax = plt.subplot(111, projection='polar')
    ax.plot(theta, r, color='r', linewidth=3)
    ax.set_rmax((int(max(amp)/100)+1)*100)
    ax.grid(True)
    ax.set_title("50Hz AC test polar plot [DUT,REF,ERR*10000]", loc = 'left',va='top',fontsize=15)
plot_polar(amp_60Hz_off_board,angle_60Hz_off_board)
```

```python
amp_50Hz_off_board = [166.8,167.88,273.42]
angle_50Hz_off_board =[192.07, 12.2, 12.58]
plot_polar(amp_50Hz_off_board,angle_50Hz_off_board)
```

```python
amp_60Hz_PCB = [166.022,166.95,264.77]
angle_60Hz_PCB =[360-14.91, 360-182.53, 360-2.324]
plot_polar(amp_60Hz_PCB,angle_60Hz_PCB)
```
