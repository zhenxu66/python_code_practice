```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
r = np.arange(0, 3.0, 0.01)
theta = 2 * np.pi * r
```

```python
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r, color='r', linewidth=3)
ax.set_rmax(2.0)
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
```

```python
#amp = [159.71,277.55,87.93]
#angle =[307.1, 127.6, 128.8]
#[DUT,REF,ERR]
amp = [170.17,170.23,239.71]
angle =[360-16, 163.4, 190]

r = [0,amp[0],0,amp[1],0,amp[2]/2,0]
theta = [0,angle[0]* np.pi/180,0, angle[1]* np.pi/180,0,angle[2]* np.pi/180,0]
plt.figure(figsize=(15,15),dpi=200)

plt.rcParams.update({'font.size': 30})
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r, color='r', linewidth=3)
ax.set_rmax((int(max(amp)/100)+1)*100)
ax.grid(True)
ax.set_title("50Hz AC test polar plot [DUT,REF,ERR]", loc = 'left',va='top',fontsize=15)
```
