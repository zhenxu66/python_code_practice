```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
#from mpl_toolkits.mplot3d import Axes3D
```

# map 3D to 2D plane, [x,y,z] to value to [x,y] to value

```python
from sklearn.datasets import make_s_curve
make_s_curve(n_samples=5)
```

```python
X, y = make_s_curve(n_samples=800)
```

```python
from mpl_toolkits.mplot3d import Axes3D
ax = plt.axes(projection='3d')

ax.scatter3D(X[:, 0], X[:, 1], X[:, 2], c=y)
ax.view_init(5, -70)
```

The sklearn.decomposition module includes matrix decomposition algorithms.
Including among others PCA, NMF or ICA. 
Most of the Linear algorithms of this module can be regarded as dimensionality reduction techniques.
(information along one direction is lost)

The sklearn.manifold module implements data embedding techniques.
Including: MDS, isomap, TSNE
(Nonlinear transformers, unsupervised)

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit(X)
X_pca = pca.transform(X)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
```

```python
from sklearn.manifold import Isomap

iso = Isomap(n_neighbors=20)
iso.fit(X)
X_iso = iso.transform(X)
plt.scatter(X_iso[:, 0], X_iso[:, 1], c=y)
```

TSNE is good to preserve local structure, but not global structure

```python
from sklearn.manifold import TSNE

tsne = TSNE()
tsne.fit(X)
```

```python
X_tsne = tsne.transform(X)
```

```python
X_tsne = tsne.embedding_
```

```python
X_tsne = tsne.fit_transform(X)

plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
```

##TSNE with the digits dataset, is very effective in showing group data

```python
from sklearn.datasets import load_digits
digits = load_digits(n_class=7)
print(digits.keys())
print(np.unique(digits.target))
```

```python
X, y = digits.data, digits.target
```

```python
X_tsne = TSNE().fit_transform(X)
```

```python
plt.scatter(X_tsne[:, 0], X_tsne[:, 1])
```

```python
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
```
